```python
# List of questions, choices, and correct answers
questions = [
    {
        "question": "What is the capital of France?",
        "choices": ["1. Berlin", "2. Madrid", "3. Paris", "4. Rome"],
        "answer": "3"
    },
    {
        "question": "Which language is primarily used for web development?",
        "choices": ["1. Python", "2. JavaScript", "3. Java", "4. C++"],
        "answer": "2"
    },
    {
        "question": "What is 5 + 7?",
        "choices": ["1. 10", "2. 11", "3. 12", "4. 13"],
        "answer": "3"
    },
    {
        "question": "Who wrote the play 'Romeo and Juliet'?",
        "choices": ["1. Charles Dickens", "2. William Shakespeare", "3. Mark Twain", "4. Jane Austen"],
        "answer": "2"
    },
    {
        "question": "What is the largest planet in our solar system?",
        "choices": ["1. Earth", "2. Mars", "3. Jupiter", "4. Saturn"],
        "answer": "3"
    }
]

def ask_question(question_data):
    """Ask a single question, get the user's answer, and return if it's correct."""
    print(f"\n{question_data['question']}")
    for choice in question_data['choices']:
        print(choice)

    user_answer = input("\nEnter the number of your answer: ").strip()

    if user_answer == question_data['answer']:
        print("Correct!\n")
        return True
    else:
        print(f"Wrong! The correct answer was: {question_data['answer']}\n")
        return False

def play_quiz():
    """Run the quiz game."""
    score = 0
    total_questions = len(questions)

    print("Welcome to the Quiz Game!")
    print(f"There are {total_questions} questions in total.\n")

    for idx, question_data in enumerate(questions, 1):
        print(f"Question {idx}:")
        if ask_question(question_data):
            score += 1

    # Display the final score
    print(f"\nYour final score is {score}/{total_questions}.")
    if score == total_questions:
        print("Perfect score! Well done!")
    elif score > total_questions / 2:
        print("Good job! You did well.")
    else:
        print("Better luck next time!")

def main():
    """Main function to start the quiz game."""
    while True:
        play_quiz()
        play_again = input("\nDo you want to play again? (yes/no): ").strip().lower()
        if play_again != 'yes':
            print("Thanks for playing! Goodbye!")
            break

if __name__ == "__main__":
    main()
```
